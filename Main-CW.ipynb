```python
#normalise the data values so that the models will not give priority to higher values.

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_X = scaler.fit_transform(complete_data_X)
scaled_X[0]
```

Following conclusions taken from data preprocessing
* Dataset has 9690 images.
* Dataset has 2303 features (48x48 pixels).
* There are no null columns.
* Dataset has 10 classes.
* This is an imbalanced dataset (i.e. distribution of classes are not equal).
* We used Minmax scaler to normalize the pixel values.

```python
# reshape data and create a 10x10 subplot to visualize data

re_X = np.reshape(scaled_X, (9690, 48, 48))

plt.figure(figsize=(10, 10))
plt.subplots_adjust(wspace=0, hspace=0)

for i, re_X in enumerate(re_X[:100]):
    plt.subplot(10, 10, i+1)
    plt.imshow(re_X, cmap="binary")
    plt.axis("off")
```
