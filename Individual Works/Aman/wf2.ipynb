```python
#normalising the data values so that all the values will be within 0 and 1

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled[0]
```

### 2. Feature Selection using PCA with 0.95 information integrity

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
pca_X = pca.fit_transform(X_scaled)
```

```python
pca_X.shape
```

we reduced the dimension to 73 columns from 2304 original columns.

73 columns keep the 95% of data of the dataset.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(pca_X, y, test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape
```

Using SGDClassifier

```python
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

sgd = SGDClassifier()
dt = DecisionTreeClassifier()
lgrg = LogisticRegression()
knn = KNeighborsClassifier()
svc = SVC()
rf = RandomForestClassifier()
```

```python
sgd.fit(X_train, y_train)
dt.fit(X_train, y_train)
lgrg.fit(X_train, y_train)
knn.fit(X_train, y_train)
svc.fit(X_train, y_train)
rf.fit(X_train, y_train)
```

```python
sgd_pred = sgd.predict(X_test)
dt_pred = dt.predict(X_test)
lgrg_pred = lgrg.predict(X_test)
knn_pred = knn.predict(X_test)
svc_pred = svc.predict(X_test)
rf_pred = rf.predict(X_test)
```

```python
[sgd.score(X_test, y_test),
dt.score(X_test, y_test),
lgrg.score(X_test, y_test),
knn.score(X_test, y_test),
svc.score(X_test, y_test),
rf.score(X_test, y_test)]
```

confusion matrix

```python
from sklearn.metrics import confusion_matrix

cm_sgd = confusion_matrix(y_test, sgd_pred)
cm_dt = confusion_matrix(y_test, dt_pred)
cm_lgrg = confusion_matrix(y_test, lgrg_pred)
cm_knn = confusion_matrix(y_test, knn_pred)
cm_svc = confusion_matrix(y_test, svc_pred)
cm_rf = confusion_matrix(y_test, rf_pred)
```

```python
plt.figure(figsize=[15,10])
plt.subplot(231)
plt. title('SGD')
sns.heatmap(cm_sgd, annot=True, fmt='g')
plt.ylabel('Actual Values')
plt.xlabel('Predicted Values')
plt.subplot(232)
plt. title('DT')
sns.heatmap(cm_dt, annot=True, fmt='g')
plt.subplot(233)
plt. title('LGRG')
sns.heatmap(cm_lgrg, annot=True, fmt='g')
plt.subplot(234)
plt. title('KNN')
sns.heatmap(cm_knn, annot=True, fmt='g')
plt.subplot(235)
plt. title('SVC')
sns.heatmap(cm_svc, annot=True, fmt='g')
plt.subplot(236)
plt. title('RF')
sns.heatmap(cm_rf, annot=True, fmt='g')
```
