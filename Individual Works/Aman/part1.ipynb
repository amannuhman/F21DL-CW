### Naïve Bayes Classification and Performance Metrices

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(X_train, y_train)
gnb_pred = gnb.predict(X_test)
gnb_pred
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test,gnb_pred)
cmdisp = ConfusionMatrixDisplay(cm)
cmdisp.plot()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,gnb_pred))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, gnb_pred)
precision = precision_score(y_test, gnb_pred, average='weighted')
recall = recall_score(y_test, gnb_pred, average='weighted')
f1 = f1_score(y_test, gnb_pred, average='weighted')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```

```python
from sklearn.metrics import roc_curve, auc

fpr = dict()
tpr = dict()
roc_auc = dict()

n_classes = 10

for i in range(n_classes):
    # Treat class i as the positive class and the others as the negative class
    y_true = (y_test2 == i)
    y_score = (gnb2_pred == i)
    fpr[i], tpr[i], _ = roc_curve(y_true, y_score)
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure(figsize=(10, 5))
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i],
             label=f'class {i} - (auc = {roc_auc[i]:0.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve for each class')
plt.legend()
plt.show()
```

### Additional Analysis using Decision Tree and Random Forest Classifier
