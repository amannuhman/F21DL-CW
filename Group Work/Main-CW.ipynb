## 2. Feature Selection

### 2.1 Selecting Top 2, Top 5, Top 10 features via correlation matrix

By Applying to complete dataset;
* Top 2 ~ 50 features
* Top 5 ~ 100 features
* Top 10 ~ 200 features

```python
ros_data = pd.DataFrame(ros_x)
ros_data['class'] = ros_y
ros_data
```

```python
# takes around 4-5mins to execute

corr_data = ros_data.corr()
```

```python
corr_data['class'].sort_values(ascending=False)
```

```python
#selecting top 50 features based on correlation matrix

temp_data = pd.DataFrame(corr_data['class'].sort_values(ascending=False))
top25_posititve = temp_data.index[1:26]
top25_negative = temp_data.index[-25:]
corr_top50_features = ros_data[top25_posititve].join(ros_data[top25_negative])
corr_top50_features
```

```python
#selecting top 100 features based on correlation matrix

temp_data = pd.DataFrame(corr_data['class'].sort_values(ascending=False))
top50_posititve = temp_data.index[1:51]
top50_negative = temp_data.index[-50:]
corr_top100_features = ros_data[top50_posititve].join(ros_data[top50_negative])
corr_top100_features
```

```python
#selecting top 200 features based on correlation matrix

temp_data = pd.DataFrame(corr_data['class'].sort_values(ascending=False))
top100_posititve = temp_data.index[1:101]
top100_negative = temp_data.index[-100:]
corr_top200_features = ros_data[top100_posititve].join(ros_data[top100_negative])
corr_top200_features
```
